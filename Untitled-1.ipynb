# Import libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
from scipy.stats import linregress
```

# Read the data and plot them
They should look parabolic

```python
# These are the raw measurements

l = [     
[2.55, 0.663],
[3.0,      0.688],
[3.4,    0.78],
[3.8,   0.854],
[4.0,      0.921],
[4.4,    1.084],
[4.6,    1.18],
[5.0,    1.402],
[5.2,  1.531],
[5.6,   1.811]
]

# These are the offsets
x0 = 0.663
t0 = 2.55

# Crerate one dimensional arrays from the raw data
s = list()
t = list()
for il in l:
    t.append(il[0])
    s.append(il[1])

# Correct for the offsets
t = np.array(t) - t0
s = np.array(s) - x0

print(t)
print(s)

# Plot the data
plt.scatter(t, s)
plt.show()
```

# Function fitting

## Fit parabola with only quadratic term

Fit $s$ = $a t^2$ 

The fitted $a$ is  $ a = {gsin(\phi) \over 2 } $

Solving for g we get:

$g = {2a \over sin(\phi)}$

So fit the curve, calculate an print g

### Curve fitting

```python
# let's define the function form
def func(s, a):
    s = a * t ** 2
    return s

# Fit
a, = optimize.curve_fit(func, xdata = t, ydata = s)[0]


print(a)

g = a * 2 / 0.03
print(g)
```

### Plot the data points and the curve
It should look parabolic

```python
x = np.linspace(0, 3, 100)
y = a * x ** 2 

plt.plot(x, y) # Plot the curve
plt.scatter(t, s) # Plot the data points
plt.show()
```
